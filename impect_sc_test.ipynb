````python
# Import required libraries
import json
import pandas as pd

# Import SkillCorner API client
try:
    from skillcorner.client import SkillcornerClient
    print("✓ skillcorner API client imported successfully")
except ImportError as e:
    print(f"✗ Error importing skillcorner: {e}")
    print("Install with: pip install skillcorner")

# Import kloppy for loading tracking data
try:
    from kloppy import skillcorner as kloppy_sc
    print("✓ kloppy (SkillCorner provider) imported successfully")
except ImportError as e:
    print(f"✗ Error importing kloppy: {e}")
    print("Install with: pip install kloppy")

print("\nAll libraries imported successfully!")
````

````python
# Load SkillCorner credentials
with open('creds/skillcorner.json', 'r') as f:
    skillcorner_creds = json.load(f)

print("Credentials loaded successfully!")
print(f"SkillCorner username: {skillcorner_creds['username']}")
````

````python
# Connect to SkillCorner API
print("Connecting to SkillCorner API...")
sc_client = SkillcornerClient(
    username=skillcorner_creds['username'],
    password=skillcorner_creds['pswd']
)
print("✓ Successfully connected to SkillCorner API!")
````

````python
# Get SkillCorner competitions
print("=" * 60)
print("SKILLCORNER - Available Competitions")
print("=" * 60)
try:
    sc_competitions = sc_client.get_competitions()
    print(f"\nTotal competitions found: {len(sc_competitions)}")
    
    if sc_competitions:
        sc_competitions_df = pd.DataFrame(sc_competitions)
        print("\nCompetitions:")
        print(sc_competitions_df.head(20))
        print(f"\nColumns: {list(sc_competitions_df.columns)}")
    else:
        print("No competitions found.")
except Exception as e:
    print(f"Error fetching competitions: {e}")
    sc_competitions = []
````

````python
# Get SkillCorner matches
print("=" * 60)
print("SKILLCORNER - Available Matches")
print("=" * 60)
try:
    sc_matches = sc_client.get_matches()
    print(f"\nTotal matches found: {len(sc_matches)}")
    
    if sc_matches:
        sc_matches_df = pd.DataFrame(sc_matches)
        print("\nSample matches:")
        print(sc_matches_df.head(20))
        print(f"\nColumns: {list(sc_matches_df.columns)}")
    else:
        print("No matches found.")
except Exception as e:
    print(f"Error fetching matches: {e}")
    sc_matches = []
````

## Using Kloppy for Tracking Data

**Kloppy** is a Python library for loading football/soccer tracking data from various providers.

**Key Features:**
- Loads tracking data from SkillCorner JSON files
- Standardized data format across providers
- Works with local files or API responses
- Supports event data and tracking frames

**Usage:**
```python
from kloppy import skillcorner as kloppy_sc

# Load from files
dataset = kloppy_sc.load(
    meta_data="match_metadata.json",
    raw_data="tracking_data.json"
)

# Access data
frames = dataset.frames  # Tracking frames
events = dataset.events  # Event data (if available)
metadata = dataset.metadata  # Match metadata
```
